```python
import pandas as pd
import matplotlib.pyplot as plt
```

# EPL

```python
epl_preCOVID_file = ['http://football-data.co.uk/mmz4281/1415/E0.csv', 'http://football-data.co.uk/mmz4281/1516/E0.csv',
                'http://football-data.co.uk/mmz4281/1617/E0.csv', 'http://football-data.co.uk/mmz4281/1718/E0.csv',
                'http://football-data.co.uk/mmz4281/1819/E0.csv']
epl_COVID_file = ['http://football-data.co.uk/mmz4281/2021/E0.csv']
```

```python
epl_preCOVID_data = pd.DataFrame()  #empty df
for filename in epl_preCOVID_file:  #for loop to read all files and combine datas
    df = pd.read_csv(filename, index_col=None, header=0)
    epl_preCOVID_data = epl_preCOVID_data.append(df)
    
epl_COVID_data = pd.DataFrame()  #empty df
for filename in epl_COVID_file:
    df = pd.read_csv(filename, index_col=None, header=0)
    epl_COVID_data = epl_COVID_data.append(df)
```

```python
# get the value count of Full Time Result as a percentage value
epl_preCOVID = pd.DataFrame(epl_preCOVID_data['FTR'].value_counts(normalize=True)*100).round(2).T.rename(
    columns={'H': 'HomeTeamWin%', 'A': 'AwayTeamWin%', 'D': 'Draw%'}, index={'FTR': 'epl_preCOVID'})

epl_COVID = pd.DataFrame(epl_COVID_data['FTR'].value_counts(normalize=True)*100).round(2).T.rename(
    columns={'H': 'HomeTeamWin%', 'A': 'AwayTeamWin%', 'D': 'Draw%'}, index={'FTR': 'epl_COVID'})

epl_data = epl_preCOVID.copy().append(epl_COVID).T #append above both df and transpose
epl_data['Growth_Rate'] = (epl_data.epl_COVID - epl_data.epl_preCOVID)
epl_data
```

```python
epl_data.plot(y=["epl_preCOVID", "epl_COVID"], kind="bar", stacked=False, legend=True, figsize=(16,9), color=['khaki', 'lightsteelblue'])
plt.xticks(size = 17, rotation='horizontal')
plt.yticks(size = 17)
plt.ylim(0, 50)
L=plt.legend(prop={'size': 14})
L.get_texts()[0].set_text('Avg of Pre COVID Seasons 2014-2019')
L.get_texts()[1].set_text('COVID Season 2020-2021')
plt.ylabel('Win Percentage', size=17, color='black')
plt.title('Win percentage comparison of Home & Away teams in EPL before & during COVID',fontsize=20, color='Blue') 
plt.text(0.02, 40, 'Home Team Win: '+str(epl_data['Growth_Rate'][0].round(2))+'%', bbox=dict(facecolor='bisque', alpha=0.5), size=20, color='red')
plt.text(1.02, 42, 'Away Team Win: '+str(epl_data['Growth_Rate'][1].round(2))+'%', bbox=dict(facecolor='bisque', alpha=0.5), size=20, color='green')
plt.text(2.02, 27, 'Draw: '+str(epl_data['Growth_Rate'][2].round(2))+'%', bbox=dict(facecolor='bisque', alpha=0.5), size=20, color='black')
plt.savefig('epl_result.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
